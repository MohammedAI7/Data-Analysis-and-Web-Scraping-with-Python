```python
import requests
from bs4 import BeautifulSoup

url = 'https://en.wikipedia.org/wiki/List_of_largest_companies_in_India'
page = requests.get(url)

soup = BeautifulSoup(page.text, "html.parser")

#print(soup.title.text)  # Just to test
#print(soup)
```

```python
soup.find_all('table')[0]
```

```python
soup.find('table', class_ = 'wikitable sortable')
```

```python
table = soup.find_all('table')[0]
```

```python
print(table)
```

```python
world_title =table.find_all('th')
```

```python
print(world_title)
```

```python
world_table_titles = [title.text.strip() for title in world_title]
```

```python
print(world_table_titles)
```

```python
df = pd.DataFrame(columns = world_table_titles)
df
```

```python
df.rename(columns={'Forbes 2000 rank': 'Forbes rank','Headquarters':'city'}, inplace=True)
```

```python
column_data =table.find_all('tr')
```

```python
df = pd.DataFrame(columns=df.columns)  # reset df to empty before inserting
for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    
    length = len(df)
    df.loc[length] = individual_row_data
```
